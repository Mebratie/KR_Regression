```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import csv
import pandas as pd
import itertools
import math
import os
import sympy as sp
import random
from sklearn.model_selection import train_test_split
from sympy import symbols, Function, diff
from mpl_toolkits.mplot3d import Axes3D
output_directory = r'../../../System/Implicitisation/Whitney_umbrella'
output_directory1 = r'../../../Data/Implicitisation/Whitney_umbrella'
output_directory2 = r'../../../results/Implicitisation/Whitney_umbrella'
def parametric_equations(u, v):
    x = u
    y = u * v
    z = v**2
    h = 0 
    return x, y, z, h
def write_csv(file_path, fieldnames, data):
    with open(file_path, 'w', newline='') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()
        for row in data:
            writer.writerow(row)
def split_data(data, split_ratio=0.8):
    split_index = int(len(data) * split_ratio)
    train_data = data[:split_index]
    test_data = data[split_index:]
    return train_data, test_data
def generate_data(u_values, v_values):
    data = []
    for u in u_values:
        for v in v_values:
            x, y, z, h = parametric_equations(u, v)
            data.append({'x1': x, 'x2': y, 'x3': z, 'trajectory': h})
    data.append({'x1': 1.0, 'x2': 1.0, 'x3': -1.0, 'trajectory': 1.0})
    return data
def plot_3d_scatter(data, output_directory):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(data['x1'], data['x2'], data['x3'])
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    plt.savefig(os.path.join(output_directory2, 'plot.png'))
    plt.close()
def main():
    u_values = np.linspace(-1, 1, 10)
    v_values = np.linspace(-1, 1, 10)
    file_name = '50.csv'
    file_path = os.path.join(output_directory, file_name)
    fieldnames = ['x1', 'x2', 'x3', 'trajectory']
    data = generate_data(u_values, v_values)
    train_data, test_data = split_data(data, split_ratio=0.8)
    train_file_path = os.path.join(output_directory1, 'trainingp_data50_' + file_name)
    test_file_path = os.path.join(output_directory1, 'holdoutp_data50_' + file_name)
    write_csv(file_path, fieldnames, data)
    write_csv(train_file_path, fieldnames, train_data)
    write_csv(test_file_path, fieldnames, test_data)
    loaded_data = np.genfromtxt(file_path, delimiter=',', names=True)
    plot_3d_scatter(loaded_data, output_directory2)
if __name__ == "__main__":
    main()
```
